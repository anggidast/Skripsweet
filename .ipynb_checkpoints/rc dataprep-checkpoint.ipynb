# Library Importing

```python
import pandas as pd
```

# Rubber Compound Database Importing (Worksheet Merging)

```python
df1 = pd.concat(pd.read_excel('JANUARI2019.xlsx', sheet_name=None), ignore_index=True)
df2 = pd.concat(pd.read_excel('FEBRUARI2019.xlsx', sheet_name=None), ignore_index=True)
df3 = pd.concat(pd.read_excel('MARET2019.xlsx', sheet_name=None), ignore_index=True)
df4 = pd.concat(pd.read_excel('APRIL2019.xlsx', sheet_name=None), ignore_index=True)
df5 = pd.concat(pd.read_excel('MEI2019.xlsx', sheet_name=None), ignore_index=True)
df6 = pd.concat(pd.read_excel('JUNI2019.xlsx', sheet_name=None), ignore_index=True)
df7 = pd.concat(pd.read_excel('JULI2019.xlsx', sheet_name=None), ignore_index=True)
df8 = pd.concat(pd.read_excel('AGUSTUS2019.xlsx', sheet_name=None), ignore_index=True)
df9 = pd.concat(pd.read_excel('SEPTEMBER2019.xlsx', sheet_name=None), ignore_index=True)
df10 = pd.concat(pd.read_excel('OKTOBER2019.xlsx', sheet_name=None), ignore_index=True)
df11 = pd.concat(pd.read_excel('NOVEMBER2019.xlsx', sheet_name=None), ignore_index=True)
df12 = pd.concat(pd.read_excel('DESEMBER2019.xlsx', sheet_name=None), ignore_index=True)
df13 = pd.concat(pd.read_excel('JANUARI2020.xlsx', sheet_name=None), ignore_index=True)
df14 = pd.concat(pd.read_excel('FEBRUARI2020.xlsx', sheet_name=None), ignore_index=True)
df15 = pd.concat(pd.read_excel('MARET2020.xlsx', sheet_name=None), ignore_index=True)
```

# Rubber Compound Database Merging

```python
frames = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15]
df = pd.concat(frames, ignore_index=True)
df
```

# Data Cleansing

```python
df = df.drop(df.loc[:,'Unnamed: 2':'Unnamed: 12'].head(0).columns, axis=1)
df = df.drop(columns=['PT. IRC INOAC INDONESIA', 'Unnamed: 17'])
df
```

```python
df = df.loc[df['Unnamed: 1']=="E-1774"]
df = df.reset_index(drop=True)
df
```

```python
df = df.drop(columns=['Unnamed: 1'])
df = df.reset_index(drop=True)
df.columns = ['hardness', 'tensile', 'elongation', 'sg']
df
```

```python
df = df[(df.hardness > 66) & (df.hardness < 74)]
df
```

```python
df = df[(df.tensile > 99)]
df
```

```python
df = df[(df.elongation > 299)]
df
```

```python
df = df[(df.sg > 1.22) & (df.sg < 1.30)]
df
```

```python
df = df.reset_index(drop=True)
df
```

# Dataset Exporting

```python
df.to_excel("rubbercompound.xlsx")
```
