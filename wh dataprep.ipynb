# Water Hose Database Merging

```python
frames = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15]
df = pd.concat(frames, ignore_index=True)
df
```

# Data Cleansing

```python
df = df.drop(df.loc[:,'Unnamed: 2':'Unnamed: 4'].head(0).columns, axis=1)
df = df.drop(columns=['PT. IRC INOAC INDONESIA', 'Unnamed: 7'])
df
```

```python
df = df.loc[df['Unnamed: 1']=="F-300"]
df = df.drop(columns=['Unnamed: 1'])
df = df.reset_index(drop=True)
df
```
